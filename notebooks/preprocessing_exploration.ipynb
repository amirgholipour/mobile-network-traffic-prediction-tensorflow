## Internet traffic

```python
internet_data_path = "data/sms-call-internet-mi-2013-11-01.txt"

data = pd.read_csv(internet_data_path, delimiter="\t", header=None, 
            usecols=[0, 1, 7], names=["square_id", "time", "internet"])
```

```python
print(len(data))
data.head(n=8)
```

```python
print("number of rows with no internet data: {0}".format(data.isna().internet.sum()))

cleaned_data = data.dropna()
# cleaned_data.head(n=8)
```

```python
sorted_data = cleaned_data.sort_values(by="time")
# sorted_data.head(n=8)
```

```python
milano_grid_size = (100, 100)

def milano_grid_mapping(cell_id):
    # returns row_nr, column_nr in a 2D grid
    num_rows, num_columns = milano_grid_size
    cell_indx = cell_id - 1
    row_nr = num_rows - 1 - cell_indx // num_rows
    column_nr = cell_indx % num_rows
    
    return row_nr, column_nr

assert milano_grid_mapping(1) == (99, 0)
assert milano_grid_mapping(15) == (99, 14)

assert milano_grid_mapping(100) == (99, 99)
assert milano_grid_mapping(101) == (98, 0)

assert milano_grid_mapping(200) == (98,99)
assert milano_grid_mapping(201) == (97, 0)

assert milano_grid_mapping(9801) == (1, 0)
assert milano_grid_mapping(9999) == (0, 98)

print("seems to work")
```

```python
groups = sorted_data.groupby("time")
previous = 0
for time, group in groups:
    assert time >= previous
    previous = time
    for index, row in group.iterrows():
        print(row)
    break
    
print("groups are sorted in time")
```
